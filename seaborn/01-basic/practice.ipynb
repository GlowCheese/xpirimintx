```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
print(*sns.get_dataset_names())
```

```python
df = sns.load_dataset("flights")
df.head(10)
```

```python
sns.lineplot(df, x="year", y="passengers")
plt.show()
```

```python
df5860 = df[(1958 <= df["year"]) & (df["year"] <= 1960)]
sns.lineplot(df5860, x="month", y="passengers", hue="year")
plt.show()
```

```python
dfjuly = df[df["month"] == "Jul"]
sns.lineplot(dfjuly, x="year", y="passengers")
plt.show()
```

```python
df5060 = df[(1950 <= df["year"]) & (df["year"] <= 1960)]
sns.boxplot(df5060, x="year", y="passengers")
plt.show()
```
