```python
# import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# tips is a sample dataset
df: pd.DataFrame = sns.load_dataset("tips")
```

```python
print(df.head(3))
```

---

```python
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.show()
```

---

```python
sns.lineplot(x="size", y="tip", data=df)
plt.show()
```

keywords:
- confidence interval (CI=95%)
- standard deviation
- dispersed / concentrated

---

```python
sns.barplot(x="size", y="tip", data=df)
plt.show()
```

---

```python
sns.histplot(x="total_bill", data=df, bins=20, kde=True)
plt.show()
```

keywords:
- bins: number of bars
- kde (Kernel Density Estimate):
    - data's distribution shape
    - spotting trends like peaks or skew
