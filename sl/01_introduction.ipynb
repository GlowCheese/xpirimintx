### 📚 Lesson 1: Introduction to Supervised Learning – "How do we teach machines with examples?"

---

#### ❓ Today's question

"How can we make a computer learn from labeled examples to make predictions on new data?"

---

#### 💡 Key ideas

* **Supervised Learning** = học có giám sát. Nghĩa là: mình đưa cho máy 1 tập data có input **X** và label **Y**, rồi bảo: "Ê, học đi để lần sau gặp input mới thì biết đoán Y."
* Bản chất: tìm 1 cái **mapping function f: X → Y**.
* Có 2 loại chính:

  1. **Classification** → Y là discrete (cat/dog, spam/not spam).
  2. **Regression** → Y là continuous (giá nhà, nhiệt độ, điểm số).
* Chiến lược:

  1. Thu thập data (X, Y).
  2. Chia train/test.
  3. Chọn model (Linear Regression, Decision Tree, Neural Net…).
  4. Train = tìm parameters để minimize loss.
  5. Evaluate trên test.

👉 Đây là nền móng: cả supervised learning chỉ xoay quanh việc **học từ data có nhãn** để đoán nhãn mới.
