#### ❓ Today's question

"Thay vì đo đạo hàm bằng cách chênh nhỏ (finite difference), làm sao tính **gradient analytically** để update parameters chính xác hơn?"

---

#### 💡 Key ideas

* Finite difference = approximate gradient, hơi ngu, tốn time, không chính xác.
* **Analytic derivative** = dùng công thức đạo hàm thật sự của loss function theo parameters.
* Ví dụ: Linear Regression 1D, MSE loss:

$$L(w, b) = \frac{1}{n}\sum_{i=1}^{n} (y_i - (w x_i + b))^2$$

* Đạo hàm theo w và b:

$$
\frac{\partial }{\partial b} (y - (wx + b))^2 = -2 (y - (wx + b))
\quad \longrightarrow \quad
\frac{\partial L}{\partial b} = -\frac{2}{n} \sum_i (y_i - (w x_i + b))
$$

$$
\frac{\partial }{\partial w} (y - (wx + b))^2 = -2x (y - (wx + b))
\quad \longrightarrow \quad
\frac{\partial L}{\partial w} = -\frac{2}{n} \sum_i x_i (y_i - (w x_i + b))
$$

* Lợi ích: nhanh, chính xác, dễ scale sang nhiều biến (multi-variable).

---

#### 📝 Practice Problems

* Rewrite Linear Regression 1D **using analytic gradient** thay vì finite difference.
* X, Y vẫn như cũ:

    ```python
    X = [1, 2, 3, 4, 5]
    Y = [2, 4, 6, 8, 10]
    ```

* Alpha = 0.01, epochs = 1000.
* In w, b cuối cùng và loss.

👉 Mục tiêu: w ≈ 2, b ≈ 0, loss ≈ 0.

---

#### 🧩 Solution

````python
import numpy as np
from tqdm import tqdm
````

````python
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 6, 8, 10])
````

````python
def loss(w: float, b: float):
    return ((Y - (w * X + b)) ** 2).sum() / len(X)


def deriv_b(w: float, b: float):
    return -2 * (Y - (w * X + b)).sum() / len(X)


def deriv_w(w: float, b: float):
    return -2 * (X * (Y - (w * X + b))).sum() / len(X)
````

````python
alpha = 0.01
epochs = 1000
````

````python
w = b = 0

for _ in tqdm(range(epochs)):
    w -= alpha * deriv_w(w, b)
    b -= alpha * deriv_b(w, b)

print(f"Final result: w={w:.4f}, b={b:.4f}, loss={loss(w, b):.4f}")
````
