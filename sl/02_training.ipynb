---

#### ❓ Today's question

"How does the machine update itself from examples to make better predictions?"

---

#### 💡 Key ideas

* **Training** = quá trình cho model nhìn data, tính toán error (loss), rồi chỉnh parameters để error nhỏ dần.
* Key steps:

  1. **Model (hypothesis)**: 1 function giả định (ví dụ Linear Regression: y = w·x + b).
  2. **Loss function**: đo sai số (e.g. Mean Squared Error cho regression).
  3. **Optimization**: tìm parameters (w, b) sao cho loss nhỏ nhất → dùng Gradient Descent.
* Gradient Descent:

  * Tính đạo hàm loss theo w, b.
  * Cập nhật ngược hướng đạo hàm:

    ```
    w = w - α * dL/dw  
    b = b - α * dL/db
    ```
  * α = learning rate (quyết định bước nhảy).

👉 Đây là tim của supervised learning: **học = minimize loss function qua optimization**.

---

#### 📝 Practice Problems

Viết Python để implement **Linear Regression 1D** bằng Gradient Descent từ scratch.

Yêu cầu:

* Input: tập data đơn giản, ví dụ:

  ```python
  X = [1, 2, 3, 4, 5]
  Y = [2, 4, 6, 8, 10]   # y = 2x
  ```
* Model: y\_pred = w \* x + b
* Loss: Mean Squared Error (MSE).
* Update rule: gradient descent.
* Chạy training 1000 epochs, in ra w, b cuối cùng và loss.

Kết quả mong đợi: w ≈ 2, b ≈ 0, loss ≈ gần 0.

---

#### 🧩 Solution

````python
X = [1, 2, 3, 4, 5]
Y = [2, 4, 6, 8, 10]
````

````python
def loss(w: float, b: float):
    res = 0
    for i in range(len(X)):
        p = w * X[i] + b
        res += (p - Y[i]) ** 2
    return res
````

````python
alpha = 0.005
epsilon = 0.005
iterations = 1000
````

````python
w = b = 0

loss_records = []

for iter in range(iterations):
    l = loss(w, b)
    loss_records.append(l)
    lw = (loss(w + epsilon, b) - l) / epsilon
    lb = (loss(w, b + epsilon) - l) / epsilon
    w -= alpha * lw
    b -= alpha * lb
````

````python
print(f"Final result: w={w:.4f}, b={b:.4f}, mse={loss_records[-1]:.4f}")
````
