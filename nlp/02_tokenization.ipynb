### 📚 Lesson 2: Tokenization – How do we split text into useful pieces?

---

#### ❓ Today's question

"How can we break down raw text into smaller units (tokens) that a computer can process?"

---

#### 💡 Key ideas

* Text gốc chỉ là chuỗi ký tự → phải **tách** ra thành đơn vị nhỏ (tokens).
* **Word-level tokenization**: đơn giản nhất, tách theo dấu cách (`"I love NLP"` → `["I", "love", "NLP"]`).
* **Problems with word-level**:

  * "NLP" vs "nlp" vs "Nlp" → case sensitivity.
  * Dấu câu (punctuation) → `"hello!"` và `"hello"` khác nhau.
  * Morphology: `"love"` vs `"loves"` vs `"loving"`.
* **Subword tokenization** (Byte-Pair Encoding, WordPiece, SentencePiece): cắt từ thành chunks nhỏ hơn → xử lý từ mới/hiếm. Ví dụ `"unhappiness"` → `["un", "happiness"]` hoặc thậm chí `["un", "##happy", "##ness"]`.
* **Character-level tokenization**: mỗi ký tự = token. Hiếm khi dùng một mình, nhưng hữu ích cho ngôn ngữ phức tạp hoặc khi dữ liệu ít.

👉 Tóm lại: Tokenization = bước đầu tiên, quyết định độ "hiểu" của mô hình với text.

---

#### 📝 Practice Problems

Viết hàm Python `tokenize(sentence)` để:

* Lowercase toàn bộ câu.
* Bỏ dấu câu (.,!?).
* Split theo khoảng trắng.

Ví dụ:

```python
sentence = "I love NLP, do you love it?"
tokenize(sentence)
# Output: ["i", "love", "nlp", "do", "you", "love", "it"]
```

---

#### 🧩 Solution

````python
def tokenize(sentence: str):
    return [
        "".join(c for c in word if c not in ".,!?")
        for word in sentence.strip().lower().split()
    ]
````

````python
sentence = "I love NLP, do you love it?"
tokenize(sentence)
````
