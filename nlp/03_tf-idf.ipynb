### 📚 Lesson 3: Bag-of-Words vs TF-IDF – How can we make word counts smarter?

#### ❓ Today's question

"If Bag-of-Words just counts words, how can we reduce the impact of common words (like 'the', 'is') so that rare but important words stand out?"

---

#### 💡 Key ideas

* **Bag-of-Words (BoW):** mỗi văn bản = vector đếm số lần xuất hiện từ. Nhưng:

  * Từ phổ biến như “the”, “and” luôn có count cao, dù chẳng mang nhiều nghĩa.
* **TF-IDF (Term Frequency – Inverse Document Frequency):** cải tiến BoW bằng cách:

  * **TF (term frequency):** tần suất từ trong văn bản (giống BoW).
  * **IDF (inverse document frequency):** phạt những từ xuất hiện trong quá nhiều văn bản.

    * Công thức:

      $$
      \text{IDF}(t) = \log \frac{N}{1 + df(t)}
      $$

      Trong đó:

      * $N$ = số văn bản.
      * $df(t)$ = số văn bản chứa từ $t$.
  * Cuối cùng:

    $$
    TF\text{-}IDF(t, d) = TF(t, d) \times IDF(t)
    $$

👉 Ý tưởng: từ nào xuất hiện nhiều trong **ít văn bản** → trọng số cao (quan trọng). Từ nào xuất hiện ở **mọi văn bản** → trọng số thấp (ít thông tin).

   * Hint:

#### 🧩 Solution

````python
from math import log
````

````python
corpus = [
    "I love NLP",
    "NLP loves me",
    "I love machine learning"
]
````

````python
vocabs = list(sorted(set(word for sent in corpus for word in sent.split())))
````

````python
df = {word: sum(1 if word in sent.split() else 0 for sent in corpus) for word in vocabs}
````

````python
idf = {word: log(len(corpus) / (1 + df[word])) for word in vocabs}
````

````python
embeddings = []
for sent in corpus:
    emb = []
    for word in vocabs:
        tf = sent.split().count(word)
        emb.append(tf * idf[word])
    embeddings.append(emb)
````
