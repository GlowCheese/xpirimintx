### 📚 Lesson 1: Text Representation – How do we turn words into numbers?

---

#### ❓ Today's question

"How can a computer represent text (words/sentences) in a way it can understand and process?"

---

#### 💡 Key ideas

* Máy tính chỉ hiểu **số** (vectors, matrices), chứ không hiểu chữ cái A, B, C.
* Vì thế, NLP phải biến đổi **text → numerical representation**.
* Các chiến lược phổ biến:

  1. **Tokenization**: cắt câu thành các đơn vị nhỏ (tokens) → có thể là word-level, subword, hay character-level.
  2. **One-hot encoding**: mỗi từ = vector toàn số 0, chỉ 1 vị trí là 1. Rất sparse, không capture meaning.
  3. **Bag-of-Words (BoW)**: đếm số lần xuất hiện của từ trong văn bản. Đơn giản nhưng mất thông tin ngữ cảnh.
  4. **TF-IDF (Term Frequency - Inverse Document Frequency)**: mở rộng BoW bằng cách giảm trọng số mấy từ xuất hiện quá phổ biến (như "the", "and").
  5. **Word Embeddings (Word2Vec, GloVe)**: ánh xạ từ → vector dense, có tính chất ngữ nghĩa (vd: vec("king") - vec("man") + vec("woman") ≈ vec("queen")).

👉 Đây là nền tảng: muốn NLP thì phải biết **biểu diễn text**.

---

#### 📝 Practice Problems

Viết code để biến list sau thành dạng Bag-of-Words vectors:

```python
corpus = [
      "I love NLP",
      "NLP loves me",
      "I love machine learning"
]
```

Output mong muốn: một ma trận (list of lists) biểu diễn BoW cho từng câu.

---

#### 🧩 Solution

````python
corpus = ["I love NLP", "NLP loves me", "I love machine learning"]
````

````python
vocabs = set()
word_idx = {}
for sent in corpus:
    for word in sent.split():
        if word not in vocabs:
            word_idx[word] = len(vocabs)
            vocabs.add(word)
````

````python
embeddings = []
for sent in corpus:
    emb = [0] * len(vocabs)
    for word in sent.split():
        emb[word_idx[word]] += 1
    embeddings.append(emb)
````
