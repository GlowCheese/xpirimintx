### 📚 Lesson 4: Word Embeddings – How can computers understand meaning between words?

---

**Mày đã học gì rồi?**

* Mày đã biết: **biểu diễn text cơ bản** (one-hot, BoW, TF-IDF) và **tokenization**.
* Hiểu nhược điểm one-hot, word-level tokenization, và TF-IDF (dù cải tiến nhưng chưa nắm ngữ nghĩa).
* Tự code được BoW và TF-IDF chuẩn.

---

#### ❓ Today's question

"If BoW/TF-IDF only know word counts, how can we make the computer capture actual **semantic similarity** between words (e.g. king \~ queen, car \~ vehicle)?"

---

#### 💡 Key ideas

* **BoW/TF-IDF**: chỉ dựa trên frequency, không có notion “từ nào gần nghĩa với từ nào”.
* **Word embeddings** (Word2Vec, GloVe): ánh xạ từ thành **vector dense, low-dimensional**, trong đó quan hệ ngữ nghĩa được giữ.

  * Ví dụ:

    $$
    \text{vec}(king) - \text{vec}(man) + \text{vec}(woman) \approx \text{vec}(queen)
    $$

* Ưu điểm:

  * Vector ngắn gọn hơn TF-IDF (100–300 dims vs hàng chục nghìn).
  * Capture similarity & analogies.
* Hạn chế:

  * Static embeddings → một từ có nhiều nghĩa thì không phân biệt được (vd: “bank” = sông vs ngân hàng).

👉 Đây là bước nền để hiểu các mô hình NLP hiện đại (BERT, GPT).
