   * Dòng thứ 1 (tất cả cột).

2. Code: Với array `a = np.arange(10)`, hãy:

   * In ra các phần tử chẵn.
   * In ra các phần tử tại index 2, 4, 6, 8.

```python
a = np.array([np.arange(4*i, 4*i+4) for i in range(4)])
print(a[1, 2])
print(a[1])
print(a[:, 3])
print(a[:2, :2])
```

```python
a = np.arange(10)
print(a[a % 2 == 0])
print(a[[2, 4, 6, 8]])
```
