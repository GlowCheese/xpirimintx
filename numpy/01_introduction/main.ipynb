#### ❓ Today's Question

**"NumPy là gì, và tại sao dev nào cũng cần dùng nó thay vì list thường của Python?"**

---

#### 💡 Key ideas

* **NumPy** = **Numerical Python** → thư viện xử lý số liệu, tối ưu cho **mảng nhiều chiều (ndarray)**.
* Tại sao không dùng list thường?

  * Python list chậm, không tối ưu cho toán học.
  * NumPy arrays: chạy bằng C dưới nền → nhanh hơn nhiều.
  * Có rất nhiều **vectorized operations** (cộng, trừ, nhân ma trận...) mà list thường phải loop.
* Core của NumPy = **ndarray object** (một dạng mảng nhiều chiều, đồng nhất về kiểu dữ liệu).

---

#### 📝 Practice Problems

Viết code tạo một array `[1, 2, 3, 4, 5]` bằng NumPy. In ra:

   * array
   * kiểu dữ liệu của array (type)
   * kiểu dữ liệu phần tử (dtype)

---

#### 🧩 Solution

```python
import numpy as np
```

```python
a = np.array([1, 2, 3, 4, 5])
print(a)
print(type(a))
print(a.dtype)
```
