## 📚 Lesson 1: Giới thiệu SciPy – Nó là gì và dùng để làm gì?

---

### ❓ Today's question

"**What is SciPy, and why is it important for scientific computing?**"

---

### 💡 Key ideas

* SciPy là **thư viện mở rộng của NumPy**. Nó cung cấp các **hàm nâng cao cho khoa học, kỹ thuật, tối ưu hóa, tích phân, xử lý tín hiệu, thống kê**.
* Nếu NumPy là vũ khí cơ bản (array, linear algebra), SciPy là **mỏ vàng** cho mày muốn giải các bài toán **nâng cao, thực tế hơn**.
* Các module quan trọng:

  * `scipy.optimize` → tối ưu hóa, tìm cực trị.
  * `scipy.integrate` → tích phân, ODE.
  * `scipy.linalg` → giải các bài toán tuyến tính nâng cao (linear system, eigenvalues).
  * `scipy.stats` → thống kê, phân phối xác suất.
  * `scipy.signal` → xử lý tín hiệu.

👉 SciPy + NumPy = combo chuẩn để làm **scientific computing**.

---

### 🧩 Solution


```python
import scipy
print(scipy.__version__)
```

```python
from scipy.integrate import quad

f = lambda x: x*x
result, err = quad(f, 0, 1)

print(result)
```
