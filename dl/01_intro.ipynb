### 📚 Lesson 1: What is Deep Learning?

---

**Mày đã học gì rồi?**

Chưa học gì cả. Đây là **bài mở màn**. Bắt đầu từ con số 0, tao sẽ đặt nền cho mày hiểu Deep Learning là cái thể loại gì.

---

#### ❓ Today's question

“Deep Learning dùng để giải quyết **vấn đề gì** trong thế giới thực?”

---

#### 💡 Key ideas

* Máy tính bình thường chỉ giỏi tính toán 1+1=2, nhưng lại gà trong việc **nhìn ảnh**, **nghe tiếng nói**, **hiểu ngôn ngữ**, …
* Deep Learning là một nhánh của **Machine Learning** chuyên xử lý những vấn đề phức tạp này nhờ **Neural Networks** (mạng nơ-ron nhân tạo).
* Ý tưởng:

  * Thay vì lập trình chi tiết từng luật → ta **cho máy học trực tiếp từ data**.
  * Data càng nhiều, mô hình càng sâu, máy càng “thông minh” hơn.
* Vấn đề DL thường xử lý:

  * **Computer Vision** (nhận diện ảnh, xe tự lái).
  * **NLP** (dịch máy, chatbot, GPT mà mày đang chửi đây).
  * **Speech** (nhận diện giọng nói, trợ lý ảo).
  * **Game / Decision making** (AlphaGo, game bot).

👉 Deep Learning là “vũ khí hạng nặng” khi data lớn, mô hình phức tạp, và muốn máy tính **tự tìm quy luật thay vì mình code tay**.
