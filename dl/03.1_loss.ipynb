#### ❓ Today's question

“Model biết mình sai hay đúng kiểu gì?”

---

#### 💡 Key ideas

* **Loss Function** = thước đo độ lệch giữa **prediction** và **ground truth**.
* Mục tiêu training: **minimize loss**.

Ví dụ:

* Regression: dùng **Mean Squared Error (MSE)**

  $$
  L = \frac{1}{n} \sum (y_{true} - y_{pred})^2
  $$
* Classification: dùng **Cross-Entropy Loss**

  $$
  L = - \sum y_{true} \cdot \log(y_{pred})
  $$

👉 Không có loss function thì model chả biết đường mà cải thiện.
