### 📚 Lesson 2: Neural Networks cơ bản – Cái “não giả” vận hành thế nào?

---

**Mày đã học gì rồi?**

* Mày biết Deep Learning chuyên xử lý **pattern phức tạp** trong data (ảnh, text, speech).
* Nó khác với rule-based vì máy **học luật từ data**, không phải con người viết từng dòng.

---

#### ❓ Today's question

“Neural Network là cái quái gì, và nó học từ dữ liệu bằng cách nào?”

---

#### 💡 Key ideas

* **Neuron nhân tạo**: Lấy cảm hứng từ não người. Mỗi neuron = 1 phép tính:

  $$
  y = f(w \cdot x + b)
  $$

  * $x$: input (dữ liệu vào)
  * $w$: weight (trọng số, mức độ quan trọng của input)
  * $b$: bias (dịch chuyển)
  * $f$: activation function (tạo phi tuyến tính → giúp mô hình học pattern phức tạp).

* **Layer**: nhiều neuron ghép lại → một lớp.

  * Input layer: nhận dữ liệu gốc.
  * Hidden layer(s): học representation.
  * Output layer: cho ra dự đoán.

* **Training**:

  1. Input data → output dự đoán.
  2. So sánh với ground truth → tính loss.
  3. Dùng **backpropagation + gradient descent** để cập nhật weight cho tốt hơn.

👉 Ý tưởng: mạng nơ-ron chỉ là 1 “máy biến đổi” dữ liệu qua nhiều lớp, tối ưu weight sao cho output gần đúng target.

---

#### 📝 Practice Problems

1. **Knowledge check:**
   Trong công thức $y = f(w \cdot x + b)$, giải thích ngắn gọn ý nghĩa của từng thành phần $w, b, f$.

2. **Application thinking:**
   Nếu không có activation function (tức là $f(x) = x$), thì mạng nơ-ron sẽ gặp vấn đề gì?

---

#### 🧩 Solution


**Question 1:**

* $w$ = weight → độ quan trọng của input.
* $b$ = bias → giúp dịch chuyển output, tránh bị kẹt tại gốc.
* $f$ = activation function → tạo phi tuyến, học pattern phức tạp.

**Question 2:**

- Nếu $f(x) = x$, tất cả layer chỉ cộng dồn
- cả network chỉ là **hàm tuyến tính to đùng**, dù có chồng 100 layer cũng như 1 layer.
