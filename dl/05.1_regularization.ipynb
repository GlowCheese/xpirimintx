### 📚 Lesson 5.1: Regularization – How do we prevent overfitting?

---

**Mày đã học gì rồi?**

* Mày biết **NN architecture** cơ bản: input → hidden layers (ReLU/ELU/Leaky ReLU/Softmax).
* Mày biết cách train bằng **loss function** (MSE, CrossEntropy), update bằng **SGD/optimizer**.
* Activation functions → xử lý vanishing gradient, dead neurons.

---

#### ❓ Today's question

"Neural networks rất mạnh, nhưng nếu model nhớ hết training data thì nó sẽ **overfit**. Làm sao để tránh overfitting và giúp model generalize tốt hơn?"

---

#### 💡 Key ideas

Có 3 cách **regularization** chính:

1. **Weight Decay (L2 regularization):**

   * Thêm penalty cho trọng số lớn.
   * Loss = Original loss + λ‖W‖²
   * Ý tưởng: ép weights nhỏ → model ít phức tạp, tránh học noise.

2. **Dropout:**

   * Trong training, tắt ngẫu nhiên một số neuron (tỉ lệ p).
   * Buộc model học nhiều đường dự phòng, không phụ thuộc 1 neuron.

3. **Early Stopping:**

   * Stop training khi loss trên validation bắt đầu tăng (nghĩa là model đang overfit).

👉 Tổng quát: regularization = ép model **“bớt thông minh”** để nó học được patterns tổng quát, không nhớ vặt.
