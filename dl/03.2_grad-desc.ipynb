#### 💡 Key ideas

* Muốn giảm loss → phải **cập nhật weight**.
* Chiêu: **Gradient Descent**.

Ý tưởng:

1. Tính gradient của loss theo weight $\frac{\partial L}{\partial w}$.
2. Update:

   $$
   w := w - \eta \cdot \frac{\partial L}{\partial w}
   $$

   * $\eta$: learning rate (tốc độ học).

👉 Gradient Descent giống như **đi xuống dốc núi để tìm chỗ thấp nhất (minimum)**.

2. **Coding:**
   Tự code bằng numpy: tìm **w** tối ưu cho bài toán linear regression $y = 2x$.

   Gợi ý data:

   ```python
   import numpy as np

   X = np.array([1, 2, 3, 4, 5])
   Y = np.array([2, 4, 6, 8, 10])
   ```

   * Khởi tạo w = 0
   * Dùng Gradient Descent để train cho tới khi w ≈ 2
   * Loss = MSE

#### 🧩 Solution


**Question 1:**

- Big $\eta$ -> harder to converges
- Small $\eta$ -> progress very slowly

**Question 2:**

````python
import numpy as np
from tqdm import tqdm
````

````python
X = np.arange(1, 6)
Y = np.arange(2, 11, 2)
````

````python
w = 0
lr = 0.001
epochs = 1000
````

````python
def loss():
    return ((Y - w*X) ** 2).mean()

def grad():
    return -2 * (X * (Y - w*X)).mean()
````

````python
for iter in tqdm(range(epochs)):
    w -= lr*grad()
````

````python
print(f"L = {loss():.4f}, w = {w:.4f}")
````
