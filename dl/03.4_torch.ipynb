#### 📝 Practice Exercise

👉 Viết lại bài **XOR** bằng PyTorch.


**Expected Output:**

* Loss giảm dần → khoảng `0.001` hoặc nhỏ hơn.
* Predictions gần giống `[0, 1, 1, 0]`.

---

#### 🧩 Solution


```python
import torch
import torch.nn as nn
import torch.optim as optim

from tqdm import tqdm
```

```python
# Data
X = torch.tensor([[0.,0.],[0.,1.],[1.,0.],[1.,1.]])
Y = torch.tensor([[0.],[1.],[1.],[0.]])
```

```python
# Training Parameters
lr = 0.1
epochs = 10000
```

```python
# Model: 2-layer NN
model = nn.Sequential(
    nn.Linear(2, 2),   # input -> hidden
    nn.Tanh(),
    nn.Linear(2, 1),   # hidden -> output
    nn.Sigmoid()
)
```

```python
# Loss + Optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)
```

```python
# Training
for epoch in tqdm(range(epochs)):
    y_pred = model(X)
    loss = criterion(y_pred, Y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
print("Final loss:", loss.item())
print(model(X).detach().numpy().T[0])
```
